```python
import xarray as xr
import gcsfs
import numpy as np
from regridding import regrid

fs = gcsfs.GCSFileSystem(token="anon")

store = fs.get_mapper('gs://weatherbench2/datasets/era5/1959-2023_01_10-wb13-6h-1440x721_with_derived_variables.zarr')
ds = xr.open_zarr(store=store, consolidated=True)

df_era5 = ds.total_precipitation_6hr.sel({'time':np.arange('2018-01-01','2023-01-11',np.timedelta64(1,'D'),dtype='datetime64[ns]')})

lons_reg = ## give lon values e.g., for IMERG 0.1 degree grid: np.arange(19.15, 54.35, 0.1)
lats_reg = ## give lat values e.g., for IMERG 0.1 degree grid:  np.arange(-13.65, 24.75, 0.1)

df_era5['longitude'] = df_era5.longitude.values - 180

regridder = regrid(lons_reg, lats_reg, df_era5)

df_era5_regridded = regridder(df_era5, keep_attrs=True)


```
